## 3. Step 4: creating loyalty flag

```python
# first: creating max_order column
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# checking results
ords_prods_merged.head(10)
```

```python
# Creating loyalty flag using loc
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking the results
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# testing the results of the new column in first 60 entries of the dataframe
ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## 4. Step 5

```python
# finding the average amount different loyalty groups spend on products
ords_prods_merged.groupby('loyalty_flag').agg({'prices':['mean']})
```

##### Surprisingly, the average price of products that loyal 
customers buy is less than the rest of the customers.

## 5. Step 6; Creating Spending Flag

```python
# Creating spending flag
# first: creating average_price column using groupby
ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# checking the new column
ords_prods_merged.head(10)
```

```python
# creating spending flag using loc
ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
#checking the frequencies of the spending flag
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
# testing the results in first 10 entries of the dataframe
ords_prods_merged[['region', 'spending_flag']].head(10)
```

## 6. Step 7; creating order frequency flag

```python
# creating average_price column using groupby
ords_prods_merged['median_days_last_order'] = ords_prods_merged.groupby(['user_id'])['last_order_in_days'].transform(np.median)
```

```python
# checking the new column
ords_prods_merged.head(10)
```

```python
# creating order frequency flag using loc
ords_prods_merged.loc[ords_prods_merged['median_days_last_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days_last_order'] <= 20) & (ords_prods_merged['median_days_last_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days_last_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```

```python
# testing the results of the new column in first 10 entries of the dataframe
ords_prods_merged[['user_id', 'order_frequency_flag', 'median_days_last_order']].head(10)
```

## 7. Exporting Dataframe

```python
# Exporting dataframe as pkl
ords_prods_merged.to_pickle(os.path.join(path,'Prepared Data', 'aggregated_data.pkl'))
```
