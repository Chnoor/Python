```python
# Import libraries

import pandas as pd
import numpy as np
import os
```

## 1. Create data to experiment on

```python
# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}
```

```python
# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

## 2. convert the dictionaries into dataframe

```python
# Convert the dictionary into dataframe
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

## 3. Concatenate the dataframes

```python
# concatenating the dataframes
# framing the dataframes: 
frames = [df, df_1]
```

```python
# concatenationg the frames:
df_concat = pd.concat(frames)
```

```python
df_concat
```

```python
# concatenationg dataframes side by side
df_concat = pd.concat(frames, axis = 1)
```

## 5. Merging Dataframes

```python
# Merging dataframes: identifier: customer_id 
df_merged = df.merge(df_1, on = 'customer_id')
```

```python
df_merged
```

```python
# Merging dataframes: identifier: customer_id and month
df_merged = df.merge(df_1, on = ['customer_id','month'])
```

```python
df_merged
```

```python
# check if dataframes match
df_merged = df.merge(df_1, on = ['customer_id'], indicator = True)
```
